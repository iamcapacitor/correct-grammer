```python
from collections import deque
class EventProcessingSystem:
    def __init__(self):
        self.event_queue = deque()
    # Add an Event
    def add_event(self, event):
        self.event_queue.append(event)
        print(f"‚úÖ Event '{event}' added to the queue.")
    # Process Next Event
    def process_event(self):
        if self.event_queue:
            event = self.event_queue.popleft()
            print(f"‚ö° Processing event: {event}")
        else:
            print("‚ö† No events to process.")
    # Display Pending Events
    def display_events(self):
        if self.event_queue:
            print("üìå Pending Events:", list(self.event_queue))
        else:
            print("‚ö† No pending events.")
    # Cancel an Event
    def cancel_event(self, event):
        if event in self.event_queue:
            self.event_queue.remove(event)
            print(f"‚ùå Event '{event}' canceled.")
        else:
            print(f"‚ö† Event '{event}' not found.")
# -------- Menu Driven Program --------
if __name__ == "__main__":
    eps = EventProcessingSystem()
    while True:
        print("\n---- Event Processing System ----")
        print("1. Add Event")
        print("2. Process Next Event")
        print("3. Display Pending Events")
        print("4. Cancel Event")
        print("5. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            event = input("Enter event name: ")
            eps.add_event(event)
        elif choice == "2":
            eps.process_event()
        elif choice == "3":
            eps.display_events()
        elif choice == "4":
            event = input("Enter event name to cancel: ")
            eps.cancel_event(event)
        elif choice == "5":
            print("Exiting... üëã")
            break
        else:
            print("‚ö† Invalid choice. Try again.")
```

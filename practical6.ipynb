```python
from collections import deque

class CallCenterQueue:
    def __init__(self):
        self.queue = deque()
    # Add a new call
    def addCall(self, customerID, callTime):
        self.queue.append((customerID, callTime))
        print(f"üìû Call from Customer {customerID} at {callTime} mins added to queue.")
    # Answer the next call
    def answerCall(self):
        if self.queue:
            customerID, callTime = self.queue.popleft()
            print(f"‚úÖ Answering call from Customer {customerID}, waited {callTime} mins.")
        else:
            print("‚ö† No calls to answer.")
    # View all calls in queue
    def viewQueue(self):
        if self.queue:
            print("üìã Pending Calls in Queue:")
            for i, (cid, ctime) in enumerate(self.queue, 1):
                print(f"{i}. Customer {cid} - Call Time {ctime} mins")
        else:
            print("‚ö† Queue is empty.")
    # Check if queue is empty
    def isQueueEmpty(self):
        return len(self.queue) == 0

# -------- Menu Driven Simulation --------
if __name__ == "__main__":
    ccq = CallCenterQueue()
    while True:
        print("\n---- Call Center Queue ----")
        print("1. Add Call")
        print("2. Answer Call")
        print("3. View Queue")
        print("4. Check if Queue is Empty")
        print("5. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            cid = input("Enter Customer ID: ")
            time = input("Enter Call Time (in minutes): ")
            ccq.addCall(cid, time)
        elif choice == "2":
            ccq.answerCall()
        elif choice == "3":
            ccq.viewQueue()
        elif choice == "4":
            if ccq.isQueueEmpty():
                print("‚úÖ Queue is empty.")
            else:
                print("‚ùå Queue has pending calls.")
        elif choice == "5":
            print("Exiting... üëã")
            break
        else:
            print("‚ö† Invalid choice. Try again.")
```

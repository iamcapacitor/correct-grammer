```python
# Selection Sort implementation
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

# Bubble Sort implementation
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
# Example employee salaries
salaries = [55000.50, 42000.75, 60000.00, 75000.25, 
            58000.10, 82000.00, 47000.90, 91000.30, 50000.45]
# Selection Sort
selection_sorted = selection_sort(salaries.copy())
print("Salaries after Selection Sort (Ascending):", selection_sorted)
print("Top 5 salaries (Selection Sort):", selection_sorted[-5:][::-1])
# Bubble Sort
bubble_sorted = bubble_sort(salaries.copy())
print("\nSalaries after Bubble Sort (Ascending):", bubble_sorted)
print("Top 5 salaries (Bubble Sort):", bubble_sorted[-5:][::-1])
```
